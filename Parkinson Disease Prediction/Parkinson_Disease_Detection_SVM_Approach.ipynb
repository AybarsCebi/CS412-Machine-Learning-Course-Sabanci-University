```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection from Kaggle and Analysis

```python
parkinson_data=pd.read_csv("/content/parkinsons.csv")
print(parkinson_data.head())
```

```python
print(parkinson_data.shape)
parkinson_data.info
```

```python
#checking for missing values on each column
parkinson_data.isnull().sum()
```

```python
#getting some statistical measureas about the data
parkinson_data.describe()
```

```python
parkinson_data['status'].value_counts()
# 1 --> Parkinson Disease
# 0 --> Healthy
```

```python
x=parkinson_data.drop(columns=['name', 'status'], axis=1)
y=parkinson_data['status']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=2)

print(x.shape, x_train.shape, x_test.shape)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
print(x_train)
```

Vector Machine Model

```python
model=svm.SVC(kernel='linear')
# training the SVM model with training data
model.fit(x_train, y_train)
```

Accuracy Score

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train, x_train_prediction)
print("Accuracy Score of the Training Data: ",training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, x_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
